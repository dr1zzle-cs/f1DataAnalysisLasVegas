We will be looking forward to analyzing the data of GP to see if we can find out the correlation with the data and the decision to box or not box. And we will also look forward to visualizing the data.

Importing Important modules for the working. I imported all the libraries by Default but you can only import the fasftf1 library as well

```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
from matplotlib.collections import LineCollection
from matplotlib import cm
import numpy as np
import pandas as pd
```

```python
# Setup plotting
plotting.setup_mpl()

# Enable the cache
ff1.Cache.enable_cache('../content/cache')

# Get rid of some pandas warnings that are not relevant for us at the moment
pd.options.mode.chained_assignment = None
```

Loading the Data


```python
# Load the session data
race = ff1.get_session(2023, 'LasVegas', 'Q')
race.load()
```

```python
df_driver = race.laps
df_driver
```

Getting the sector speeds

```python
dff = df_driver.filter(regex = "Sector\dTime|Driver$")
dff
```

**Data Preprocessing**

```python
dff = dff.groupby("Driver").min()
dff
```

```python
best_driver_sector = '|'.join(dff.idxmin())
dff.filter(regex = best_driver_sector, axis = 0)
```

```python
best_minimum = dff.min()
best_minimum
```

```python
dff
```

```python
dff = dff.sub(best_minimum)
dff
```

```python
dff = dff.apply(lambda x : x.dt.total_seconds())
dff
```

```python
dff = dff.melt(ignore_index = False)
dff
```

```python
dff.variable = dff.variable.replace('.*(\d).*', r'S\1', regex = True)
dff
```

Sorting the result

```python
dff = dff.reset_index()
order_driver = dff.groupby(['Driver']).value.sum().sort_values().index[::-1]
dff = dff.set_index('Driver').loc[order_driver]
dff
```

Gap to best Lap

```python
best_minimum_lap = best_minimum.sum().total_seconds()

best_minimum = (df_driver
          .query('IsPersonalBest == True')
          .groupby('Driver').LapTime.min()
          .dt.total_seconds()
          .sub(best_minimum_lap)
        )

best_minimum_gap = best_minimum.loc[order_driver]
best_minimum_gap
```

# Data Visualization



```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
fig = px.bar(data_frame = dff, y = dff.index, x = 'value', color = 'variable')
fig.update_layout(template = 'plotly_dark')

fig.add_trace(
  go.Scatter(
      x = best_minimum_gap.values, y = best_minimum_gap.index,
      mode = 'lines+markers',
      name = 'Personal Best Lap', line = dict(
          dash = 'dashdot', width = 2, color = '#00ff00'),
      marker = dict(size = 5)
  )
)

scaler = 50
scaler_margin = 1
fig.update_layout(
    xaxis_title = F'Gap (s) to Overall Ideal Lap [{best_minimum_lap} s]',
    legend_title_text = 'Sectors and Overall',
    title = '2023 - Las Vegas GP Qualifying',
    width = 16*scaler, height = 9*scaler,
)

fig.add_annotation(
    xref = 'paper',
    yref = 'paper',
    x = 1.15,
    y = 0.2,
    text = 'datons',
    showarrow = False,
    xanchor = 'right',
    yanchor = 'bottom',
    align = 'right',
    font = dict(color = 'grey')
)
```
